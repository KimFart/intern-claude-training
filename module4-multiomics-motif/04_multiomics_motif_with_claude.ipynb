### **# Question 4-3** <a name="issue-4_3"></a>
MEME uses expectation-maximization (EM) to find overrepresented sequence patterns. The algorithm iterates between estimating motif occurrence probabilities (E-step) and updating the position weight matrix (M-step). E-value measures how often a motif at least this good would appear by chance — low E-value (< 0.05) means the motif is unlikely to be random. MEME takes DNA sequences (FASTA format) as input, not coordinates.

```bash
meme input_sequences.fasta -dna -mod zoops -nmotifs 5 -minw 6 -maxw 20 -oc meme_output/
```
